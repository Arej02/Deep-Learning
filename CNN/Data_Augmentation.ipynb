```python
import tensorflow as tf
from tensorflow import keras
from keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pylab as plt
```

```python
dog=image.load_img('Dog.png',target_size=(200,200))
```

```python
plt.imshow(dog)

plt.show()
```

```python
datagen = ImageDataGenerator(
        rotation_range=40,
        width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        fill_mode='nearest')
```

```python
dog=image.img_to_array(dog)
```

```python
dog.shape
```

```python
input_batch=dog.reshape(1,200,200,3) # So we are sending 1 image of size 200*200*3
```

```python
i=0
for output in datagen.flow(input_batch,batch_size=1,save_to_dir="aug"):
    i=i+1
    if i==10:
        break

# So this will create 10 new images
```

```python
import pandas as pd
import matplotlib.pylab as plt
import tensorflow as tf

from tensorflow.keras import Sequential,layers
from tensorflow.keras.layers import Dense,Conv2D,MaxPooling2D,Flatten
```

```python
def create_alexnet(input_shape=(227,227,3),num_classes=1000):
    alexNet=Sequential()
    alexNet.add(layers.Conv2D(96,(11,11),strides=(4,4),input_shape=input_shape,activation="relu"))
    alexNet.add(MaxPooling2D((2,2),strides=(2,2)))
    alexNet.add(layers.Conv2D(256,(5,5),padding="same",activation="relu"))
    alexNet.add(layers.MaxPooling2D((2,2),strides=(2,2)))
    alexNet.add(layers.Conv2D(384,(3,3),padding="same",activation="relu"))
    alexNet.add(layers.Conv2D(384,(3,3),padding="same",activation="relu"))
    alexNet.add(layers.Conv2D(256,(3,3),padding="same",activation="relu"))
    alexNet.add(MaxPooling2D((3,3),strides=(2,2)))

    alexNet.add(layers.Flatten())

    alexNet.add(layers.Dense(4096,activation='relu'))
    alexNet.add(layers.Dropout(0.5))
    alexNet.add(layers.Dense(4096,activation='relu'))
    alexNet.add(layers.Dropout(0.5))
    alexNet.add(layers.Dense(num_classes,activation='softmax'))

    return alexNet



model=create_alexnet()

model.compile(loss="categorical_crossentropy",optimizer="Adam")

model.summary()
```

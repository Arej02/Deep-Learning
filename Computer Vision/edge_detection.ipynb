```python
import pandas as pd
import numpy as np 
import cv2
# ! pip install ipykernel
```

```python
image=cv2.imread("Dog.jpg")
h,w,c=image.shape
print(h,w,c)
```

```python
grey_image=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```python
# Canny edge detection: ( This will detect the edges)
edge=cv2.Canny(grey_image,100,200) # 100 is the lower threshold and 200 is the upper threshold


cv2.imshow("canny",edge)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Thresholding: (This will make the image complete black and white)
ret,thresh_img=cv2.threshold(grey_image,120,225,cv2.THRESH_BINARY) 


cv2.imshow("thresholding",thresh_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Bitwise Operation:

img1=np.zeros((512,512), dtype=np.uint8) # The height and width of both the images should be same respectively
img2=np.zeros((512,512), dtype=np.uint8) # Use balck and white image only

cv2.circle(img1,(150,150),100,255,-1)
cv2.rectangle(img2,pt1=(100,100),pt2=(250,250),thickness=-1,color=255)


bitwise_and=cv2.bitwise_and(img1,img2) # Intersection
bitwise_or=cv2.bitwise_or(img1,img2) # Union
bitwise_not=cv2.bitwise_not(img1) # Black becomes white and white becomes black
```

```python
cv2.imshow("window",bitwise_and)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cv2.imshow("window",bitwise_or)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cv2.imshow("window",bitwise_not)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

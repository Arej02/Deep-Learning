```python
import pandas as pd
import numpy as np
import cv2
```

```python
dog=cv2.imread("Dog.png")
```

```python
h,w,c=dog.shape
print(f"Height:{h}, Width;{w}, Number of channel:{c}")
```

```python
# Resize the image:
resized_image=cv2.resize(dog,(h//2,w//2))
```

```python
cv2.imshow("Original Image",dog)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
cv2.imshow("Resized Image",resized_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Crop the image:
cropped_image=dog[400:,250:]
```

```python
cv2.imshow("Cropped Image",cropped_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Rotate the image:
M=cv2.getRotationMatrix2D((w//2,h//2),45,1.0)
rotated=cv2.warpAffine(dog,M,(w,h))
```

```python
cv2.imshow("Rotated Image",rotated)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Flip the image:
flip=cv2.flip(dog,0) # Verticle: Top to Bottom
flip2=cv2.flip(dog,1) # Horizontal: Left to Right
flip3=cv2.flip(dog,-1) # Both ways

cv2.imshow("Flipped Image",flip3)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Draw a line:
line=cv2.line(dog,pt1=(0,0),pt2=(800,517),thickness=2,color=(255,0,0))

cv2.imshow("Line",line)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Draw a rectangle:
rectangle=cv2.rectangle(dog,pt1=(0,0),pt2=(30,60),thickness=-1,color=(255,0,0))

cv2.imshow("Rectangle",rectangle)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Draw a circle:
circle=cv2.circle(dog,center=(250,400),radius=50,thickness=2,color=(255,40,50))

cv2.imshow("Circle",circle)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Put a text
text_image = cv2.putText(
    dog,  # image
    "Hi I am a dog!",  # text
    (90, 90),  # org (origin/position) - Note: (x, y) coordinates
    cv2.FONT_HERSHEY_COMPLEX,  # fontFace - MUST be positional
    fontScale=1.2,  # font scale
    color=(0, 0, 255),  # BGR color (Blue=0, Green=0, Red=255)
    thickness=2  # thickness in pixels
)

cv2.imshow("text", text_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

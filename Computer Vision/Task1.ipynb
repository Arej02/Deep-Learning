```python
import pandas as pd
import numpy as np
import cv2
```

```python
# Take the path from the user and open the image
# Convert it into greyscale and show /save it depending on the users choice/path.
path_image=input("Enter the path of the image:")
try:
    image=cv2.imread(path_image)

    if image is None:
        raise FileNotFoundError("Could not load the image")
except FileNotFoundError:
    print("The file is not found")
else:
    h,w,c=image.shape
    print(f"The height is:{h}")
    print(f"The width is:{w}")
    print(f"The number of channels is:{c}")

    grey_image=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
    save_image=input("Do you want to save the grey scale image or just display it? (Y/N)?; Y to save")

    if save_image.lower()=="y":
        path=input("Enter the path:")
        cv2.imwrite(path,grey_image)

    else:
        cv2.imshow("window",grey_image)
        cv2.waitkey(0)
        cv2.destroyAllWindows()
```

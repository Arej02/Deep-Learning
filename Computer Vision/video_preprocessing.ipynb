```python
import pandas as pd
import cv2
```

```python
# Showing the live video of the webcame: 
cap=cv2.VideoCapture(0)
```

```python
while True:
    ret,frames=cap.read() # It returns true or false
    cv2.imshow("webcam",frames)

    if cv2.waitkey(1) & 0xff==ord('x'): # If the user enters x it breaks and the video turns off
        break

cap.release()
cv2.destroyAllWindows()
```

```python
# To Save the video:
cap=cv2.VideoCapture(0)
frame_width=int(cv2.get(cv2.CAP_PROP_FRAME_WIDTH)) # Get the video's width 
frame_height=int(cv2.get(cv2.CAP_PROP_FRAME_HEIGHT)) # Get the video's height
codec=cv2.VideoWriter(*'XVID')
recorded=cv2.VideoWriter("my_video.avi",codec,20,(frame_width,frame_height))
```

```python
while True:
    ret,frames=cap.read() # It returns true or false
    cv2.write(recorded)
    cv2.imshow("webcam",frames)

    if cv2.waitkey(1) & 0xff==ord('x'): # If the user enters x it breaks and the video turns off
        break

cap.release()
recorded.release()
cv2.destroyAllWindows()
```

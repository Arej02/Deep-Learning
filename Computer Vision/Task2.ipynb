```python
import pandas as pd
import numpy as np
import cv2
```

```python
# Ask the user for the location of the file:
# Ask the user what shape do they want to draw: (Line/circle/rectangle/text)
# Then ask the required stuffs
# Then ask if they want to save it or show:
```

```python

try:
    path=input("Enter the path of the file:")

    if path is None:
        print("The file is empty")

except FileNotFoundError:
    raise ("The file doesnot exist in the given path")

else:
    image=cv2.imread(path)
    h,w,c=image.shape
    shape=input("Enter the shape that you want to draw (Circle,Rectangle,Line or just text)")

    
    if shape.lower().strip()=="circle":
        img=cv2.circle(image,(w//2,h//2),100,(255,0,0),-1)
    elif shape.lower().strip()=="line":
        img=cv2.line(image,(w//2,h//2),(w,h),(255,0,0),-1)
    elif shape.lower().strip()=="rectangle":
        img=cv2.circle(image,(w//2,h//2),(w,h),(255,0,0),-1)
    elif shape.lower().strip()=="text":
        text=input("Enter the text:")
        img = cv2.putText(
            image,  
            text,  
            (90, 90),  
            cv2.FONT_HERSHEY_COMPLEX,  
            fontScale=1.2,  
            color=(0, 0, 255),  
            thickness=2 
        )

    choice=input("Do you want to save the image or show it? Y to show")

    if choice.lower().strip()=="y":
        cv2.imshow("window",img)

        cv2.waitKey(0)
        cv2.destroyAllWindows()

    elif choice.lower().strip()=="n":
        cv2.imwrite("Save.png",img)

    else:
        print("Enter the correct option")

        
```

```python
import pandas as pd
import numpy as np
import cv2
```

```python
# Read the image of the dog:
dog=cv2.imread("Dog.png")
```

```python
# Shape of the image:
h,w,c=dog.shape
print(f"Height:{h}")
print(f"Width:{w}")
print(f"Channel:{c}")
```

```python
# Display the image of the dog:
cv2.imshow("window",dog)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Convert the image into grayscale:
grey_image=cv2.cvtColor(dog,cv2.COLOR_BGR2GRAY)
```

```python
# Save the image:
cv2.imwrite("grey_dog.png",grey_image)
```

```python
# Blur the image:
blur_image=cv2.GaussianBlur(dog,(5,5),0)

cv2.imshow("window",blur_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Show the edges:
edge_detection=cv2.Canny(dog,100,200)

cv2.imshow("window",edge_detection)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

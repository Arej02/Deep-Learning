```python
import pandas as pd
import numpy as np
import cv2
```

```python
# Gaussian Blurring:
try:
    image=cv2.imread("Dog.jpg")

    if image is None:
        raise FileNotFoundError("Could not load the image")
except FileNotFoundError:
    print("The file is not found")
else:
    h,w,c=image.shape
    print(f"The height is:{h}")
    print(f"The width is:{w}")
    print(f"The number of channels is:{c}")

    gaussian=cv2.GaussianBlur(image,(21,21),0) # The higher the kernel value the more the blur

    cv2.imshow("window",gaussian)
    cv2.waitKey(0)
    cv2.destroyAllWindows()


    
```

```python
# Median Blurring:
try:
    image=cv2.imread("Dog.jpg")

    if image is None:
        raise FileNotFoundError("Could not load the image")
except FileNotFoundError:
    print("The file is not found")
else:
    h,w,c=image.shape
    print(f"The height is:{h}")
    print(f"The width is:{w}")
    print(f"The number of channels is:{c}")

    median=cv2.medianBlur(image,3) 

    cv2.imshow("window",median)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
# Sharpen Blurring:
try:
    image=cv2.imread("Dog.jpg")

    if image is None:
        raise FileNotFoundError("Could not load the image")
except FileNotFoundError:
    print("The file is not found")
else:
    h,w,c=image.shape
    print(f"The height is:{h}")
    print(f"The width is:{w}")
    print(f"The number of channels is:{c}")

    sharpen_kernel=np.array([
        [0,-1,0],
        [-1,5,-1],
        [0,-1,0]
    ])

    sharpen=cv2.filter2D(src=image,ddepth=-1,kernel=sharpen_kernel) 

    cv2.imshow("window",sharpen)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```
